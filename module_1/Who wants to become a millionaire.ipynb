Вопрос 18
Какой актер снялся в большем количестве высокобюджетных фильмов? Примечание: в фильмах, где бюджет выше среднего по данной выборке.

- Mark Wahlberg
- Matt Damon
- Angelina Jolie
- Adam Sandler

```python
data=pd.read_csv('data.csv') # считали csv-файл в датафрейм
mean_budget=data.budget.mean() #средний бюджет
#создали датафрейм актеров с индексом идентификатора фильма
actors=pd.DataFrame(data['cast'].str.split('|').tolist(),index=data['imdb_id']).stack().reset_index([0])
#создали датафрейм бюджетов с индексом идентификатора фильма
a_profit=pd.DataFrame(data['budget'].tolist(),index=data['imdb_id']).stack().reset_index([0])

#объединение слева актеров и бюджетов по индексу идентификатора фильма
a_data = actors.merge(a_profit, on='imdb_id', how='left')
a_data.columns = ['id','actor','budget'] #переименование столбцов датафрейма
#посчитали кол-во актеров в отфильтрованном по бюджету датафрейме
display(a_data[a_data.budget>mean_budget].actor.value_counts())

answer_ls.append("Matt Damon") #записали ответ
answer_ls
```

Вопрос 22
Какая студия заработала больше всего денег в жанре комедий за все время?

Warner Bros
Universal Pictures (Universal)
Columbia Pictures
Paramount Pictures
Walt Disney

```python
data=pd.read_csv('data.csv') # считали csv-файл в датафрейм
data['net_income'] = data['revenue']-data['budget'] #добавили колонку с прибылью в датафрейм
# фильтруем датафрейм по жанру комедии, и кладем компании в список, разбив по разделителю
com_company=data[data.genres.str.contains('Comedy')].production_companies.str.split('|').tolist()

#преобразуем список в датафрейм компаний с выручкой фильмов комедий
com_companies=pd.DataFrame(com_company,index=data[data.genres.str.contains('Comedy')]['revenue']).stack().reset_index([0])
com_companies.columns=['net_income','companies']
#заменяем все компании Universal* в датафрейме на Universal Pictures (Universal)
com_companies.loc[com_companies.companies.isin(['Universal','Universal Pictures']),'companies']='Universal Pictures (Universal)'

#в сгруппированном по компаниям датафрейме, отсортированном по убыванию прибыли, находим индекс максимальной прибыли
maxp_com_company=com_companies.groupby('companies').sum().sort_values('net_income', ascending=False).idxmax().net_income
display(maxp_com_company)

answer_ls.append("Universal Pictures (Universal)")
answer_ls
```

Киноиндустрия активно развивается, прибыль компаний растет с каждым годом (например, в 2014 г. индустрия заработала в 2 раза больше, чем в 2004). Поэтому прибыльность более поздних месяцев будет значительно перевешивать предыдущие, что может привести к искажению выводов.

Январь
Июнь
Декабрь
Сентябрь
Май

```python
data=pd.read_csv('data.csv') # считали csv-файл в датафрейм
data['net_income'] = data['revenue']-data['budget'] #добавили колонку с прибылью в датафрейм
#добавили в датафрейм столбец месяц, преобразованием из даты
data['Month'] = pd.to_datetime(data['release_date']).dt.month

#сводная таблица суммарной прибыли по месяцам и году
pmonth_by_year=pd.pivot_table(data, values='net_income', index='Month', columns='release_year', aggfunc='sum')
p_monthly=pmonth_by_year.idxmax().value_counts()
display(p_monthly)

answer_ls.append("Июнь")
answer_ls
```

Вопрос 32
Названия фильмов какой студии в среднем самые длинные по количеству слов?

Universal Pictures (Universal)
Warner Bros
Jim Henson Company, The
Paramount Pictures
Four By Two Productions

```python
data=pd.read_csv('data.csv') #считали данные из csv-файла в  датафрейм
words = (data.groupby('imdb_id')['original_title']  # группируем датафрейм по 
         .agg(' '.join)  # Объединяем сообщения в одну строку
         .str.replace(r'[.?,]', '')  # Удаляем ненужные символы
         .str.lower()  # Переводим в нижний регистр

         # Разбиваем строку на список слов, конвертируем в множество,
         # чтобы получить только уникальные слова и считаем длину множества
         .apply(lambda x: len(set(x.split())))
         .to_frame()  # Конвертируем в датафрейм
         .reset_index()  # Сбрасываем индекс
)
#создаем датафрейм компаний с идентификаторами фильмов
companies=pd.DataFrame(data['production_companies'].str.split('|').tolist(),index=data['imdb_id']).stack().reset_index([0])
#объединяем слева компании и кол-во слов по идентификатору фильма
words_companies = companies.merge(words, on='imdb_id', how='left')
words_companies.columns= ['id','company','words']
# сортируем по убыванию среднего кол-ва слов сгруппированный по компании датафрейм
w_company=words_companies.groupby(['company'])[['words']].mean().sort_values(['words'],ascending=False)
display(w_company)
answer_ls.append("Four By Two Productions")
answer_ls
```

Вопрос 36
У какого из режиссеров самый высокий процент фильмов со сборами выше бюджета?

Quentin Tarantino
Steven Soderbergh
Robert Rodriguez
Christopher Nolan
Clint Eastwood

```python
data=pd.read_csv('data.csv') #считали csv-файл в датафремй
data['net_income'] = data['revenue']-data['budget'] #добавили столбец прибыли в датафрейм

#создали датафрейм режиссеров и прибыли
df=pd.DataFrame(data['director'].str.split('|').tolist(), index=data['net_income']).stack().reset_index([0])
df.rename(columns={0:'director'},inplace=True)

#отфильтровали датафрейм по списку
df=df[df['director'].str.contains('Quentin Tarantino|Steven Soderbergh|Robert Rodriguez|Christopher Nolan|Clint Eastwood',na=False)].reset_index()

list=['Quentin Tarantino','Steven Soderbergh','Robert Rodriguez','Christopher Nolan','Clint Eastwood']
directors={}
#идем в цикле по датафрейму режиссеров
for i in list:
    df1=df.query('director in @i') #фильтруем датафрейм по режиссеру из списка
    df_income=df1.query('net_income>0')  #фильтруем датафрейм по прибыльным фильмам    
    percent=len(df_income)/len(df1['net_income']) #отношение кол-ва прибыльных фильмов к общему кол-ву фильмов режиссера
    directors[i]=percent
display(directors)
answer_ls.append("Quentin Tarantino")
answer_ls
```
