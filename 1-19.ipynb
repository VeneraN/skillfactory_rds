```python
for i in range(30, 42): # Условия для прохода по циклу
    # Тело цикла
    if i%2 == 0:
        print('Число', i, 'чётное')
    else:
        print('Число', i, 'нечётное')
```

```python
for i, letter in enumerate('Привет!'):
    #Здесь i - номер итерации, а letter - символ номер i из строки 'Привет!'
    print(i, letter)
```

```python
word=input()
word_new=''
for i,letter in enumerate(word):
    if i%2!=0:
        word_new=word_new+letter
print(word_new)
```

```python
year_1 =int(input())
year_2 =int(input())

for i in range(year_1,year_2+1):
    if (i%4==0 and i%100!=0) or (i%4==0 and i%400==0):
        print(i,'год високосный')
    else: print (i, 'год невисокосный')
```

```python
word=input()
word1=''
for i,letter in enumerate(word):
    if letter=='а' or letter=='я' or letter=='о' or letter=='ё' or letter=='э' or letter=='е' or letter=='у' or letter=='ю' or letter=='и' or letter=='ы':
        word1=word1+letter
print(word1)
```
